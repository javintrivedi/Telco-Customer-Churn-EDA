##EDA for churn analysis of Telco!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv.xls')
df
```

##Data Cleaning!

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df=df.dropna(subset = ["TotalCharges"])
```

```python
df = df.reset_index(drop=True)
```

##EDA Starts

```python
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Customer Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()
```

```python
sns.countplot(data=df, x='gender', hue='Churn', palette='Set2')
plt.title('Gender Distribution by Churn')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
```

```python
gender_churn = df.groupby(['gender', 'Churn']).size().unstack()

gender_churn.plot(kind='bar', stacked=False)
plt.title('Gender-wise Churn Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.legend(title='Churn', labels=['No', 'Yes'])
plt.show()
```
